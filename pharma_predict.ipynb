```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_score = pd.read_csv('/content/drive/MyDrive/DIGICHROME/cleandata1.csv')
train_score
```

We define dependent and independent variable and convert them into array

```python
# we define dependent variable
dep_var = 'Sales'

#define independent variable
indep_var =train_score.columns.drop(['Store','Promo2SinceYear','Date','Sales'])
```

```python
indep_var
```

```python
#Create the data of independent variables
X = train_score[indep_var].values

#create the dependent vaariable data
y = train_score[dep_var].values
```

```python
X
```

```python
y
```

```python
train_score[indep_var]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

splitting the dataset

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2,random_state=0)

print(X_train.shape)
print(X_test.shape)
```

Linear Regression

```python
 #scaling the x values
 scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# fitting the data into Linear Regression mode
linear_regression = LinearRegression()
linear_regression.fit(X_train,y_train)
```

```python
y_pred_li = linear_regression.predict(X_test)
y_pred_li
```

```python
linear_regression.score(X_train,y_train)
```

```python
regression_Dataframe = pd.DataFrame(zip(y_test,y_pred_li),columns = ['actual','pred'])
regression_Dataframe
```

```python
# plot for residual error

# setting plot style
plt.style.use('fivethirtyeight')

# plotting residual errors in training data
plt.scatter(linear_regression.predict(X_train),
            linear_regression.predict(X_train) - y_train,
            color="green", s=10,
            label='Train data')

# plotting residual errors in test data
plt.scatter(linear_regression.predict(X_test),
            linear_regression.predict(X_test) - y_test,
            color="blue", s=10,
            label='Test data')

# plotting line for zero residual error
plt.hlines(y=0, xmin=0, xmax=50, linewidth=2)

# plotting legend
plt.legend(loc='upper right')

# plot title
plt.title("Residual errors")

# method call for showing the plot
plt.show()
```

**Random Forest**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error , r2_score
import numpy as np
```

```python
#create a random forest regressor
random_forest = RandomForestRegressor(n_estimators=500,max_depth=8,n_jobs=2)
```

```python
#Fit the random forest to the training data
random_forest.fit(X_train,y_train)
```

```python
# Make predictions on the test data
y_pred_rf = random_forest.predict(X_test)

from sklearn.metrics import mean_absolute_error as MAE
from sklearn.metrics import mean_squared_error

mean_absolute_error_rf = MAE(y_test,y_pred_rf)
print("mean_absolute_error_rf :",mean_absolute_error_rf)

#calculate the mean square error(MSE) between the predicted and actual values
MSE_rf = mean_squared_error(y_test,y_pred_rf)
print("MSE_rf :",MSE_rf)

#calculate the RMSE
RMSE_rf = np.sqrt(MSE_rf)
print("RMSE_rf :",RMSE_rf)

#calculate the RMPSE
sales_mean = np.mean(y_test)
RMPSE_rf = RMSE_rf/sales_mean
print("RMPSE_rf :",RMPSE_rf)

#calculate r2 score
r2_rf = r2_score(y_test,y_pred_rf)
print("r2_rf :",r2_rf)
```
